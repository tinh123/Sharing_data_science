## II. Terminologies in Statistics - Statistics for Data Science
- The population is the set of sources from which data has to be collected.
- A Sample is a subset of the Population
- A Variable is any characteristics, number, or quantity that can be measured or counted. A variable may also be called a data item.

Types of analysis:
- Quantitative Analysis: Quantitative Analysis or Statistical Analysis is the science of collecting and interpreting data with numbers and graphs to identify patterns and trends.
- Qualitative Analysis: Qualitative or Non-Statistical Analysis gives generic information and uses text, sound and other forms of media to do so.


Categorical data:
- Nominal values represent discrete units and are used to label variables, that have no quantitative value => distinguish between difference objects, variable identifier and have no order. Examples: name, gender, language,...
- Ordinal values represent discrete and ordered units => the same with Nominal but Ordinal values have order and the difference between the values is not really known => order objects. Examples: ranking 10 best songs, happiness, customer satisfaction (rank 5 start,..)

Numerical data:
- Discrete data: data can only take on certain values => can't be measured but it can be counted => represents information that can be categorized into a classification.
- Continuous data:
    - Interval values represent ordered units that have the same difference => a variable that contains numeric values that are ordered and where we know the exact differences between the values. Examples: temperature (...,-2,-1,0,1,2,.. and 0 degree C is not equal 0 degree K or degree F).
    - Ratio values are also ordered units that have the same difference => Ratio values are the same as interval values but the Zero is exactlly zero.Examples: length(0 cm = 0 m = 0 km, ...), money (0 vnđ = 0 dolar). We can tranform information with a conversion factor.

## III. Categories in Statistics

### 1. Categories in statistics
There are two main categories in Statistics:
- Descriptive Statistics
- Inferential Statistics

### a. Descriptive statistics and understanding
- Descriptive Statistics uses the data to provide descriptions of the population, either through numerical calculations or graphs or tables.


- Descriptive Statistics used to helps organize data and focuses on the characteristics of data providing parameters => giving short summaries about the sample and measures of the data
- Descriptive statistics are broken down into measures of central tendency and measures of variability (spread).
    - Measures of central tendency include the mean, median, and mode.
    - Measures of variability include the standard deviation, variance, the minimum and maximum variables, and the kurtosis and skewness.


### Measures of central tendency 
- A measure of central tendency is a single value that attempts to describe a set of data by identifying the central position within that set of data => short summary statistics.
- These measures indicate where most values in a distribution fall and are also referred to as the central location of a distribution => help to fill missing values  

#### 1. Mode
- The value most recurrent in the sample set.


#### 2. Median
- Measure of the central value of the sample set, the values is ordered


#### 3. Mean
- Measure of average of all the values in a sample

### Measures of the Spread
We also have measures of the spread: Range, Inter Quartile Range (IQR), Variance, Standard Deviation.

#### Range
The range is the difference between the smallest value and the largest value in a dataset.


#### Inter Quartile Range (IQR)
Quartiles divide an ordered dataset into four equal parts, and refer to the values of the point between the quarters. A dataset may also be divided into quintiles (five equal parts) or deciles (ten equal parts).
- The lower quartile (Q1) is the point between the lowest 25% of values and the highest 75% of values. It is also called the 25th percentile.
- The second quartile (Q2) is the middle of the data set. It is also called the 50th percentile, or the median.
- The upper quartile (Q3) is the point between the lowest 75% and highest 25% of values. It is also called the 75th percentile.


Inter Quartile Range (IQR) is the measure of variability, the difference between the upper (Q3) and lower (Q1) quartiles, and describes the middle 50% of values when ordered from lowest to highest
- The interquartile range is useful in identifying when a value is an outlier.

Ref: https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51


#### The Variance and Standard deviation

- Variance describes how much a random variable differs from its expected value(mean).The Variance is defined as the average of the squared of deviations.($\frac{\sum{(x-\bar{x})^2}}{N}$)
    - Deviation is the difference between each element from the mean (x - $\bar{x}$)
    - Population Variance is the average of squared deviations.
    - Sample Variance is the average of squared differences from the mean
- The Standard Deviation is the measure of the dispersion of a set of data from its mean.. Standard Deviation is just the square root of Variance

```python

# Interquartile Range
# Import numpy library 
import numpy as np 

# First quartile (Q1) 
Q1 = np.percentile(randomlist, 25, interpolation = 'midpoint') 
  
# Third quartile (Q3) 
Q3 = np.percentile(randomlist, 75, interpolation = 'midpoint') 
  
# Interquaritle range (IQR) 
IQR = Q3 - Q1 
  
print(IQR) 
```
